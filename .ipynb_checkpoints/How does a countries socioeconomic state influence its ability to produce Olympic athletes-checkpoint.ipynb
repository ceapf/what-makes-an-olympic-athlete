# What makes an Olympic athlete?

After the 2024 Summer Olympics in paris, multiple datasets began to be released on countries that participated, as well as their placements and medallists, partitioned by event. One question I seek to answer is "How does an athletes country of origin and its socioeconomic state influence the production of said athlete in addition to the different ". There are many factors that go into making an olympic athlete, however in this analysis we hone in on things like a countries GDP and economic status, its education, and several other indicators that might influence the countries ability to produce elite athletes.

## Import Dependencies and Tools

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import dask.dataframe as dd
```

## Import Data to be Analyzed

Here we have sourced "World Development Indicators" from https://datacatalog.worldbank.org/search?q=gdp&start=0&sort= and preprocessed it in excel by creating a column of uniquely identifiable and officially declared countries (excluding regions) as well as their country code, and then transposed the different indicators as column names. Each of these uniquely identified countries and indicators have been split into 4 separate dataframes, each representing a year between 2020-2023.

```python
xls = pd.ExcelFile('country_dev_ind.xlsx')
xls2 = pd.ExcelFile('medals.xlsx')
xls3 = pd.ExcelFile('medallists.xlsx')

df2020 = pd.read_excel(xls,'2020')
df2021 = pd.read_excel(xls,'2021')
df2022 = pd.read_excel(xls,'2022') 
df2023 = pd.read_excel(xls,'2023')

goldmedals2024 = pd.read_excel(xls2,'Gold')
silvermedals2024 = pd.read_excel(xls2,'Silver')
bronzemedals2024 = pd.read_excel(xls2,'Bronze')

medallists2024 = pd.read_excel(xls3)
```

## Cleaning/Preparing the Data for Analysis

After looking over the dataset, there are multiple indicators that do not have any data present across all countries. The below code seeks to identify and remove these indicators as they will not provide anything useful in answering our overarching question.

```python
df2020_coldrop = df2020.drop(df2020.loc[:,~df2020.any()].columns,axis=1)
df2021_coldrop = df2021.drop(df2021.loc[:,~df2021.any()].columns,axis=1)
df2022_coldrop = df2022.drop(df2022.loc[:,~df2022.any()].columns,axis=1)
df2023_coldrop = df2023.drop(df2023.loc[:,~df2023.any()].columns,axis=1)

print('Number of Empty Columns Dropped for 2020: '+str(df2020.shape[1]-df2020_coldrop.shape[1]))
print('Number of Empty Columns Dropped for 2021: '+str(df2021.shape[1]-df2021_coldrop.shape[1]))
print('Number of Empty Columns Dropped for 2022: '+str(df2022.shape[1]-df2022_coldrop.shape[1]))
print('Number of Empty Columns Dropped for 2023: '+str(df2023.shape[1]-df2023_coldrop.shape[1]))
```

We can see that with each year that gets closer to the present (2024 at the time of writing this), there is less and less data available. This is most likely due to the comprehensive nature of the dataset and the indicator it attempts to capture, coupled with the difficulty of gathering the said data for each country. It makes sense that the 2020 data would be the most complete as there has been more time for the data to be gathered.

```python
def getMedals(country):
    gold = goldmedals2024.loc[goldmedals2024['Country'] == country, goldmedals2024.columns[2:]].sum(axis=1).astype(int).iloc[0]
    silver = silvermedals2024.loc[silvermedals2024['Country'] == country, silvermedals2024.columns[2:]].sum(axis=1).astype(int).iloc[0]
    bronze = bronzemedals2024.loc[bronzemedals2024['Country'] == country, bronzemedals2024.columns[2:]].sum(axis=1).astype(int).iloc[0]
    return {"Country":country,"Golds":gold,"Silvers":silver,"Bronzes":bronze,"Total":(gold+silver+bronze)}

medalslist = []
for country in df2020_coldrop['Country']:
    medalslist.append(getMedals(country))

medalsdf = pd.DataFrame(medalslist)

merged2020 = pd.merge(df2020_coldrop, pd.DataFrame(medalslist), on='Country')
merged2021 = pd.merge(df2021_coldrop, pd.DataFrame(medalslist), on='Country')
merged2022 = pd.merge(df2022_coldrop, pd.DataFrame(medalslist), on='Country')
merged2023 = pd.merge(df2023_coldrop, pd.DataFrame(medalslist), on='Country')
```

Next we wish to create a correlation matrix to identify which features might have a linear relationship with accruing medals. Due to the computational complexity of creating a correlation matrix being $O(n^2 \times m)$ where n is the number of columns and m is the number of rows, coupled with the count of columns present in our data (1,496), it would be far too computationally intensive to compare every feature in the dataset, not to mention we are solely seeking a comparison between the countries development indicators and its ability to produce Olympic athletes.

```python
merged2020corrprep = merged2020.iloc[:,2:].astype(float)
merged2021corrprep = merged2021.iloc[:,2:].astype(float)
merged2022corrprep = merged2022.iloc[:,2:].astype(float)
merged2023corrprep = merged2023.iloc[:,2:].astype(float)

lastfour = merged2020corrprep.iloc[:,-4:]
remcol = merged2020corrprep.iloc[:,:-4]
correlation_matrix = pd.DataFrame()

for col in lastfour: 
    correlation_matrix[col] = remcol.corrwith(merged2020corrprep[col])

last_column = correlation_matrix.iloc[:, -1]
correlation_matrix = correlation_matrix.iloc[:, :-1]
correlation_matrix.insert(0, last_column.name, last_column)

correlation_matrix = correlation_matrix.sort_values(by='Total', ascending=False)    

plt.figure(figsize=(40, 100))
ax = sns.heatmap(correlation_matrix, cmap='coolwarm', cbar=True)
yticks = ax.get_yticks()
ax.set_yticks(yticks + 0.5)

# Adjust y-axis and x-axis ticks
plt.yticks(ticks=range(len(correlation_matrix.index)), labels=correlation_matrix.index, rotation=0, fontsize=3)  # Increase y-axis spacing by making labels smaller
plt.xticks(rotation=90, fontsize=12)  # Bring x-axis labels closer together by rotating them

# Adjust the aspect ratio of the plot for better spacing
plt.gca().set_aspect(aspect='auto')


plt.title('Correlation Matrix Heatmap')

plt.tight_layout()
plt.savefig('correlation_matrix_heatmap.png', dpi=300)  # Save with high resolution
```

# Correlation Matrix Analysis
Now that we have created our correlation matrix with each development indicator paired with each countries ability to earn medal, we can look at which indicators have a linear relationship with medal count and which have an inversely linear relationship.

### Positive Correlation (Linear Relationship)
The below list of indicators are those that had the greatest positive correlation with a countries ability to accrue medals, meaning that as the value of these indicators increased, so did the medal count. The below elaborations on each of the top indicators aims to draw a logical conclusion as to why these indicators would have such positive correlation.

```python
correlation_matrix.head(10)
```

1. **General government final consumption expenditure (current US\$)**    
    - The indicator titled "General government final consumption expenditure (current US$)" in the World Development Indicators dataset represents the total value of the government's consumption expenditure in current U.S. dollars. This expenditure includes all government spending on goods and services that are consumed by individuals or the community as a whole.The correlation between government final consumption expenditure and Olympic success likely reflects a country's overall investment in its human capital and infrastructure, which directly impacts its ability to train and develop elite athletes. Governments that prioritize these expenditures create conditions conducive to producing athletes who can compete at the highest levels, thus accruing more medals at events like the Olympics.<br><br>
    
2. **GDP (constant 2015 US\$)**    
    - The strong correlation between GDP (constant 2015 US$) and a country's Olympic medal count can be attributed to the extensive resources available in wealthier countries to invest in sports development. This investment spans infrastructure, talent identification, training programs, and international competition, all of which are essential components of Olympic success. Countries with a high GDP have the means to support and cultivate their athletes, resulting in a higher likelihood of winning medals.<br><br>
    
3. **GNI (constant 2015 US\$)**    
    - The correlation between GNI (constant 2015 US$) and a country's ability to accrue Olympic medals can be attributed to the broader economic capabilities that GNI represents. A higher GNI indicates that a country has the economic resources to invest in all aspects of sports development, from grassroots programs to elite training facilities. This investment directly impacts a country's ability to produce world-class athletes capable of competing successfully at the Olympics.<br><br>
    
4. **Final consumption expenditure (constant 2015 US\$)**    
    - The correlation between "Final consumption expenditure (constant 2015 US$)" and a country's ability to accrue Olympic medals can be understood through the lens of economic capacity and the resultant impact on sports development. Countries with higher consumption expenditure are more likely to have the infrastructure, programs, and cultural focus necessary to support athletes, leading to greater success in international competitions like the Olympics. This expenditure reflects the overall economic environment that enables countries to invest in and prioritize athletic excellence.<br><br>
    - This appears similar to the number one most correlated indicator, and while both indicators measure consumption, "Final consumption expenditure" gives a broad picture of all spending in the economy, adjusted for inflation, whereas "General government final consumption expenditure" specifically measures government spending on public services at current prices. The former encompasses both private and public spending, providing a holistic view of economic consumption, while the latter focuses narrowly on the government's role in the economy.<br><br>
    
5. **GDP (current US\$)**    
    - The correlation between "GDP (current US\$)" and a country's ability to accrue Olympic medals is rooted in the direct relationship between economic power and the resources available for sports development. Countries with larger economies can invest more in their athletes, provide better training and facilities, and foster a culture of excellence in sports. This economic advantage translates into greater success on the Olympic stage, as wealthier nations are better equipped to develop and support world-class athletes.<br><br>
    
    - "GDP (current US\$)" is a broad indicator of economic power, but its correlation with Olympic success is likely weaker than other indicators because it does not directly reflect the specific investments in public goods and services that are critical for athletic development. Indicators like GNI, government final consumption expenditure, and inflation-adjusted measures provide more direct insights into how economic resources are allocated toward sports, education, and public health, which are crucial for nurturing Olympic talent. This specificity in measuring economic activity related to sports explains why GDP might rank lower in correlation with Olympic success compared to other economic indicators.<br><br>
    - The key difference between these two GDP measures lies in how they account for inflation. "GDP (constant 2015 US\\$)" provides a clearer and more reliable measure of real economic growth, which is essential for long-term investments like those required for Olympic success. This explains why it ranks higher in correlation with Olympic medals, as it better reflects a country’s true economic capacity to invest in sports. On the other hand, "GDP (current US$)" is more volatile and less reflective of sustained economic strength, which is why it ranks lower in correlation with Olympic success.<br><br>
    
6. **Gross National Expenditure (current US\$)**
    - The correlation between "Gross National Expenditure (current US\$)" and a country's ability to accrue Olympic medals lies in the broad economic capacity that GNE represents. High national expenditure indicates that a country has the resources to invest heavily in sports, from grassroots programs to elite training facilities. This broad investment, supported by both the public and private sectors, enhances a nation's ability to develop and support athletes who can compete successfully on the Olympic stage. While it may not be as directly targeted as government spending on sports (reflected in indicators like government final consumption expenditure), GNE provides a comprehensive picture of a nation's economic environment and its potential to support Olympic success.<br><br>
    
7. **Trademark applications, nonresident, by count**
    - While "Trademark applications, nonresident, by count" might not seem directly related to Olympic success at first glance, it reflects the broader economic and cultural dynamics that support a nation’s ability to succeed internationally. High levels of nonresident trademark applications indicate strong economic globalization, innovation, and international influence—all factors that contribute to a country’s capacity to invest in sports and develop world-class athletes. This strategic global engagement, reflected in both business and sports, can enhance a country’s ability to win Olympic medals.
   

### Negative Correlation (Inversely Linear Relationship)
The below list of indicators includes those most negatively correlated with a countries ability to accrue medals, meaning that as these value increased or decreased, medal count would inversely follow (one goes up the other goes down).

```python
correlation_matrix.tail(10).sort_values(by='Total')
```

1. **Net errors and omissions (BoP, current US\$)**
    - The strong inverse correlation between "Net errors and omissions (BoP, current US\$)" and a country’s ability to produce Olympic medals suggests that countries with large negative discrepancies in their balance of payments face economic challenges that impede their ability to invest in sports. These discrepancies may reflect broader issues like economic instability, poor governance, and capital flight, all of which can detract from a nation’s capacity to develop world-class athletes and achieve success in the Olympics. <br><br>
    
2. **Net secondary income (Net current transfers from abroad) (current US\$)**
    - The inverse correlation between "Net secondary income (Net current transfers from abroad) (current US\$)" and a country's ability to produce Olympic medals can be attributed to the economic challenges and dependencies that this indicator often reflects. Countries with high levels of net secondary income are frequently those that rely on external financial support, either from remittances or foreign aid, which may indicate underlying economic weaknesses. These weaknesses can detract from the ability to invest in the long-term development of sports infrastructure and athlete training programs, ultimately limiting the country's capacity to produce Olympic-level athletes and achieve success in international competitions.<br><br>
    
3. **Net Secondary Income (BoP, current US\\$)**
    - The inverse correlation between "Net Secondary Income (BoP, current US\\$)" and a country's ability to earn Olympic medals likely reflects economic challenges and the prioritization of different national needs in countries with high reliance on external financial transfers. These transfers, often captured in both "Net Secondary Income (BoP, current US\\$)" and "Net Secondary Income (Net current transfers from abroad) (current US\\$)," may indicate underlying economic vulnerabilities or dependencies on remittances and foreign aid. However, the key difference between these indicators is that the former is embedded within the broader context of the Balance of Payments (BoP), providing a holistic view of how these transfers impact the overall economic position of the country, whereas the latter focuses more directly on the financial inflows and outflows from abroad.<br><br>

    - In countries where net secondary income is high, particularly in the BoP context, there may be less emphasis on investing in sectors like sports that require long-term planning and stable domestic resources. This can lead to underinvestment in sports infrastructure and athlete development, limiting the country’s ability to produce Olympic-level athletes and achieve success in international competitions. Thus, countries heavily reliant on secondary income from abroad might struggle to develop the strong domestic economic foundations necessary for sustained Olympic success.<br><br>
   
4. **Portfolio investment, net (BoP, current US\\$)**
    - The inverse correlation between "Portfolio Investment, net (BoP, current US\\$)" and a country's ability to accrue Olympic medals likely reflects the economic instability and short-term focus associated with significant net portfolio investment outflows. Countries experiencing these outflows may prioritize immediate financial stability over long-term investments in sports, leading to reduced funding for athlete development and sports infrastructure. This focus on short-term financial gains, coupled with the challenges of capital flight, can limit a nation's ability to build the necessary foundation for Olympic success.

# Olympic Event Medallist Averages

### Height
Intuitively one can discern that height plays a factor in the success of Olympic athletes within specific disciplines and events. Below we separate these values out from our data, drop zeroes prior to averages to get as close as possible to determining the average height of a successful athlete within a discipline.

```python
def cm_to_feet_inches(cm):
    inches = cm / 2.54
    feet = int(inches // 12)
    remaining_inches = inches % 12
    return f"{feet}ft {remaining_inches:.2f}in"

def height_to_inches(height):
    feet, inches = height.split('ft ')
    inches = inches.replace('in', '')
    return int(feet) * 12 + float(inches)

medallists_filtered = medallists2024[medallists2024['height'] != 0]
average_heights = medallists_filtered.groupby('event')['height'].mean()
filtered_averages = average_heights[average_heights >= 120]
average_heights_feet_inches = filtered_averages.apply(cm_to_feet_inches)

average_heights_df = average_heights_feet_inches.reset_index()
discevent = medallists2024.iloc[:,8:10].drop_duplicates()

avg_heights = pd.merge(discevent, average_heights_df, on='event')

avg_heights['Height_in_inches'] = avg_heights['height'].apply(height_to_inches)

# Sort the dataframe by the new column
avg_heights_sorted = avg_heights.sort_values(by='Height_in_inches', ascending=False)

# Drop the helper column if you don't need it
avg_heights_sorted = avg_heights_sorted.drop(columns='Height_in_inches')
```
